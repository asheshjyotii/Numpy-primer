```python
import numpy as np
```

```python
#homogenous py list to numpy array
lst = [1,2,3,4]
arr = np.array(lst)
arr
```

```python
# other way of achieving the same
```

```python
arr = np.array([1,2,3,4])
arr
```

```python
# heterogenous py list to numpy arr
lst_1 = [1,2,3.0,4,5]
arr_1 = np.array(lst_1)

lst_2 = [1,2,3,4.0,'hello']
arr_2 = np.array(lst_2)

print(arr_1,'\n',arr_2)
```
